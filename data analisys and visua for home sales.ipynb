```python
plt.close()
x = data_frame["AVERAGE SALE PRICE"]
y = data_frame["LOWEST SALE PRICE"]
a = data_frame["HIGHEST SALE PRICE"]
b = data_frame["TOTAL NO. OF PROPERTIES"]
```

```python
plt.figure(figsize = (10,10))
plt.subplot(1,2,1)
plt.plot(x,y, marker="o",color ="#446699",ls="--")
plt.title("first one")
plt.xlabel("AVERAGE SALE PRICE")
plt.ylabel("LOWEST SALE PRICE")

plt.subplot(1,2,2)
plt.plot(x,a, marker = "o", color = "#227766",ls ="--")
plt.title("sec one")
plt.xlabel("AVERAGE SALE PRICE")
plt.ylabel("HIGHEST SALE PRICE")
plt.savefig("myplot.png")

plt.savefig("my plot.png")
plt.show()
```

```python
plt.close()
df2 = data_frame["TYPE OF HOME"].value_counts()
df2
```

```python
total = data_frame.shape[0]
perc = [df2.iloc[0]/total,df2.iloc[1]/total,df2.iloc[2]/total]

plt.figure(figsize=(10,5))
plt.pie(perc, labels = df2.index, startangle = 90, autopct="%1.2f%%",explode=(0.4,0.1,0.1), shadow = True)   # shado not apper becouse the background is black
plt.title("MY PIE OF PERC OF TYPE OF HOME")
plt.axis("equal")
#plt.legend()                                  # tist it

plt.savefig("my pie.png")
plt.show()

```
